```python
def check_missing(df):
    null_val = df.isnull().sum()
    percent = 100 * df.isnull().sum()/len(df)
    missing_table = pd.concat([null_val, percent], axis=1)
    col = missing_table.rename(columns = {0 : 'Num', 1 : 'Rate'})
    return col

# Display columns missing values are under 1%.
print("Data #"+str(len(full_data)))
cols = check_missing(full_data)
cols.sort_values(by="Rate", ascending=False)
```

```python
# imputing NaN with feature value suggested by dataset description
cat_df['GarageType'].fillna('NG', inplace=True)
cat_df['GarageCond'].fillna('NG', inplace=True)
cat_df['GarageFinish'].fillna('NG', inplace=True)
cat_df['GarageQual'].fillna('NG', inplace=True)
cat_df['BsmtFinType2'].fillna('NB', inplace=True)
cat_df['BsmtExposure'].fillna('NB', inplace=True)
cat_df['BsmtQual'].fillna('NB', inplace=True)
cat_df['BsmtCond'].fillna('NB', inplace=True)
cat_df['BsmtFinType1'].fillna('NB', inplace=True) 
cat_df['MasVnrType'].fillna('None', inplace=True)
cat_df['Alley'].fillna('NAc', inplace=True)
cat_df['Fence'].fillna('NF', inplace=True)
cat_df['PoolQC'].fillna('NP', inplace=True)
cat_df['FireplaceQu'].fillna('NFp', inplace=True)
cat_df['MiscFeature'].fillna('None', inplace=True)
```

```python
for v in cat_df.columns:
    cat_df[v] = cat_df[v].fillna(cat_df[v].mode()[0]) #mode imputation
    
    cat_df[v]=cat_df[v].where(cat_df[v].isin(cat_df[v].value_counts().nlargest(3).index), other='Other')
    
    cat_df = pd.get_dummies(cat_df, columns=[v], drop_first=True) #OneHotEncoding
```
